```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
tmdb = pd.read_csv("/content/tmdb_5000_movies.csv")
tmdb.head()
```

```python
sns.displot(tmdb["revenue"])
plt.title("Distribuição da receita dos filmes")
plt.show()
# O gráfico abaixo mostra a receita, em bilhoes de dólares, registrada dos filmes apresentados na lista de dados. Mostrando que pouquíssimos filmes registraram marcas altas
# de arrecadação e muitos filmes nem sequer arrecadaram ou registraram a receita.
```

```python
sns.displot(tmdb["budget"])
plt.title("Distribuição do orçamento dos filmes")
plt.show()
# Assim como no gráfico anterior, vemos colunas altas para as parcelas inferiores, nesse caso relacionado ao investimento usado no filme, esses valores não necessariamente querem dizer
# um orçamento nulo dos filmes, podem ser apenas falta de registro.
```

```python
com_faturamento = tmdb.query("revenue > 0")
sns.displot(com_faturamento["revenue"])
plt.title("Distribuição do faturamento dos filmes (registrados)")
plt.show()
# Dessa vez, o gráfico mostra os filmes que registraram os faturamentos, ainda é perceptível que poucos filmes atingiram marcas de bilhões e a maioria arrecadou valores
# muito menores e próximos de zero.
```
