```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
tmdb = pd.read_csv("/content/tmdb_5000_movies.csv")
tmdb.head()
```

```python
sns.displot(tmdb["revenue"])
plt.title("DistribuiÃ§Ã£o da receita dos filmes")
plt.show()
# O grÃ¡fico abaixo mostra a receita, em bilhoes de dÃ³lares, registrada dos filmes apresentados na lista de dados. Mostrando que pouquÃ­ssimos filmes registraram marcas altas
# de arrecadaÃ§Ã£o e muitos filmes nem sequer arrecadaram ou registraram a receita.
```

```python
sns.displot(tmdb["budget"])
plt.title("DistribuiÃ§Ã£o do orÃ§amento dos filmes")
plt.show()
# Assim como no grÃ¡fico anterior, vemos colunas altas para as parcelas inferiores, nesse caso relacionado ao investimento usado no filme, esses valores nÃ£o necessariamente querem dizer
# um orÃ§amento nulo dos filmes, podem ser apenas falta de registro.
```
